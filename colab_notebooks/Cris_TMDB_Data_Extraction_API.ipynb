# NETFLIX DATASET EXRTACTION

```python
# Import data from the webpage with the API key
netflix_json = requests.get('https://api.themoviedb.org/3/discover/movie?api_key=83f0cfd44916bbd39b2d5c5669e79ef3&with_watch_providers=8&watch_region=GB').json()
```

```python
# Retrieving information from the results section from the first page
netflix = netflix_json["results"]
```

```python
# Retrieving information from all pages by a loop
for page in range(2, netflix_json["total_pages"]+1):
    netflix_json = requests.get('https://api.themoviedb.org/3/discover/movie?api_key=83f0cfd44916bbd39b2d5c5669e79ef3&with_watch_providers=8&watch_region=GB' + f"&page={page}").json()
    netflix.extend(netflix_json["results"])
```

```python
# Convert the database into a dataframe
netflix_df = pd.DataFrame(data=netflix, columns=['title', 'id', 'vote_average', 'vote_count', 'original_language','genre_ids','popularity','release_date'])
```

```python
# Save the file to csv
netflix_df.to_csv('netflix_movies.csv', index=False)
```
